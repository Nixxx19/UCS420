```python
import numpy as np
import matplotlib.pyplot as plt

M = float(input("Enter a value for M: "))
x = np.linspace(-10, 10, 400)
y1 = M * x**2
y2 = M * np.sin(x)

plt.figure(figsize=(8, 5))
plt.plot(x, y1, label=r'$y = Mx^2$', color='b', linestyle='-')
plt.plot(x, y2, label=r'$y = Msin(x)$', color='r', linestyle='--')

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Function Plots for y = Mx² and y = Msin(x)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = {'Subject': ['Math', 'Science', 'English', 'History', 'Computers'],
        'Score': [85, 90, 78, 88, 92]}
df = pd.DataFrame(data)

plt.figure(figsize=(8, 5))
ax = sns.barplot(x="Subject", y="Score", data=df, palette="viridis")

for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width()/2, p.get_height()),
                ha='center', va='bottom', fontsize=12)

plt.title("Scores of Subjects")
plt.xlabel("Subjects")
plt.ylabel("Scores")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

roll_number = int(input("Enter your roll number: "))
np.random.seed(roll_number)
data = np.random.randn(50)
fig, axs = plt.subplots(2, 2, figsize=(10, 8))
axs[0, 0].plot(np.cumsum(data), color='b', marker='o', linestyle='-')
axs[0, 0].set_title("Cumulative Sum Plot")
axs[0, 0].grid(True)

axs[0, 1].scatter(range(50), data, color='r', alpha=0.6)
axs[0, 1].set_title("Scatter Plot with Random Noise")
axs[0, 1].grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/AnjulaMehto/MCA/main/company_sales_data.csv"
df = pd.read_csv(url)
print("Column Names:", df.columns.tolist())
df.columns = df.columns.str.strip()
print("Updated Column Names:", df.columns.tolist())

if 'Month' not in df.columns:
    print("Error: 'Month' column not found. Please check dataset structure.")
else:
    profit_column = None
    for col in df.columns:
        if 'profit' in col.lower():
            profit_column = col
            break

    if profit_column:
        df[profit_column] = pd.to_numeric(df[profit_column], errors='coerce')

        plt.figure(figsize=(8, 5))
        sns.lineplot(x=df["Month"], y=df[profit_column], marker="o", color="b")
        plt.title("Total Profit per Month")
        plt.xlabel("Month")
        plt.ylabel("Total Profit")
        plt.xticks(rotation=45)
        plt.grid(True)
        plt.show()
    else:
        print("Error: 'Total Profit' column not found.")

    plt.figure(figsize=(10, 6))
    for column in df.columns[1:-1]:
        sns.lineplot(x=df["Month"], y=df[column], label=column)

    plt.title("Sales Data for All Products")
    plt.xlabel("Month")
    plt.ylabel("Units Sold")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    plt.show()

    df.set_index("Month").plot(kind='bar', figsize=(12, 6))
    plt.title("Sales Data Across All Months")
    plt.xlabel("Month")
    plt.ylabel("Values")
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()
```
